```python
#number of days and a range
def percentFinder(industry, number_of_days):
    percentArray = []
    for i in range (len(industry)):
        openS = industry[i]['Open'].sum(axis = 0)
        close = industry[i]['Close'].sum(axis = 0)
        difference = openS - close
        decimal = difference/openS
        percent = decimal * 100
        percentArray.append(percent)
    return percentArray
```

```python
def getCompanyInDates(company, start, end):
    dates = (company['Date'] > start) & (company['Date'] <= end)
    company = company.loc[dates]
    return company 
```

```python
def clusteringData(industry_number, start, end):
    comp = companyFinder(industry_number)
    data = []
    
    for company in comp:
        dates = (company['Date'] > start) & (company['Date'] <= end)
        companyInDates = company.loc[dates]
        stdmean = companyInDates['Close'].mean() / companyInDates['Close'].std()
        data.append(stdmean)
    
    
    return data
```

```python
def dataFinder(industry_number, start, end):
    data = clusteringData(industry_number, start, end)

    from sklearn.cluster import KMeans

    b = np.reshape(data, (-1,1))
    Kmean = KMeans(n_clusters=3)
    Kmean.fit(b)
    Kmean.labels_

    one = []
    two = []
    three = []

    ret = []

    company = companyFinder(0)


    for i in range(len(Kmean.labels_)):
        if(Kmean.labels_[i] == 0):
            one.append(company[i])
        elif(Kmean.labels_[i] == 1):
            two.append(company[i])
        elif(Kmean.labels_[i] == 2):
            three.append(company[i])

    ret.append(one)
    ret.append(two)
    ret.append(three)
    
    return ret
```
